```python
# Create a series of fruits(string) of length 5 and view it
import numpy as np
import pandas as pd
fruits=np.array(["mango","chickoo","banana","cherry","kiwi"])
fruits = pd.Series(fruits)
print(fruits)
# Create a series of prices(int) of length 5 and view it
prices=np.array([100,20,20,60,120])
prices=pd.Series(prices)
print(prices)
```

```python
# Create a fruit_prices dataframe out of the 2 series you just made and view it
d={'fruits':fruits, 'price':prices}
fruits_prices=pd.DataFrame(d)
print(fruits_prices)
```

```python
# Find the mean, sum, standard deviation, variance of the prices
print('MEAN:')
print(fruits_prices.mean())
print('\nMEDIAN:')
print(fruits_prices.median())
print('\nMODE:')
print(fruits_prices.mode())
```
